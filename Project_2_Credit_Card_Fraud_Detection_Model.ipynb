Importing the libraries

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# load the dataset to the dataframe pandas
credit_card_data = pd.read_csv('/creditcard-Dataset.csv')
```

```python
# print 1st 5 rows of the dataset
credit_card_data.head()
```

```python
 credit_card_data.tail()
```

```python
# dataset info
credit_card_data.info()
```

```python
# check the no of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit trans and fraud trans
credit_card_data['Class'].value_counts()
```

This dataset is highly unbalanced.


legit trans---> 0

fraud trans---> 1

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0]
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the values for both these values of trans
credit_card_data.groupby('Class').mean()
```

The method of UNDER SAMPLING
----Build a sample dataset which contain similar distribution of legit and fraud trans

No of fraud trans is = 169

```python
legit_sample = legit.sample(n=169)
```

Concatenate two Dataframes

```python
new_dataset = pd.concat([legit_sample, fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

```python
# splitting the data into features and targets
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

```python
# split the data into training data and testing data using train_test_plit func
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=2)
```

```python
print(X.shape,X_train.shape,X_test.shape)
```

Model training

Logistic Regression

```python
model = LogisticRegression()
```

```python
# training the LR model with the training data
model.fit(X_train,Y_train)
```

Model Evaluation bsed on accuracy scifre

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction , Y_train)
```

```python
print('Accuracy on training data:',training_data_accuracy)
```

```python
# accuracy on testing data
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction , Y_test)
```

```python
print('Accuracy on testing data:',testing_data_accuracy)
```
